```python
# 1.Prepare origianl data and synthetic data 

import os
import pandas as pd
import numpy as np


os.chdir('/home/jwkim/ADR/')

data = pd.read_csv("./demo/data/demo_data.csv")
syn_data = pd.read_csv("./demo/data/demo_syn_data.csv")
```

```python
# 2. Choose the key and target variables

key = ["Gene", "Gender"]
target = "Disease"
```

```python
# 3. Calculate ADR (Default version)

import ADR

adr = ADR.ADR(data = data, 
              syn_data = syn_data, 
              key = key, 
              target = target)

print("ADR Score: ", np.round(adr.calculate(), 3))
```

```python
# 3. Calculate ADR (Detailed version)

import ADR

adr = ADR.ADR(data = data, 
              syn_data = syn_data, 
              key = key, 
              target = target,
              risk = "accuracy", 
              weight = "gini_impurity",
              imputation = "appr", 
              neighborhood = 2, 
              normalize = True, 
              show_all = True)

score, score_df = adr.calculate()
print("ADR Score: ", np.round(score, 3))
score_df
```
